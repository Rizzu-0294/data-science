```python
import matplotlib.pyplot as plt
import pandas as np
import numpy as np
import seaborn as sb
tip = sb.load_dataset('tips')
tip.head(10)
# tip[[0,1]]
plt.plot(tip['total_bill'],marker='o',color='g')
plt.title("plot the details as graphically of tabular dataframe")
plt.show()
```

```python
sb.lineplot(tip["total_bill"])
```

```python
sb.scatterplot(x="total_bill",y="tip",data=tip)
plt.show()
```

```python
sb.histplot(tip["total_bill"])
plt.show()
```

```python
plt.hist(tip["total_bill"])
plt.show()
```

```python
sb.kdeplot(tip["total_bill"])
```

```python
sb.relplot(x="total_bill",y="tip",data=tip,kind="line")
```

```python
sb.relplot(x="total_bill",y="tip",data=tip,hue="time")
plt.show()
```

```python
sb.relplot(x="total_bill",y="tip",data=tip,hue="time",style="smoker",col="sex")
plt.show()
```

```python

sb.catplot(x="day",y="total_bill",data=tip,kind="boxen")

plt.show()
```
