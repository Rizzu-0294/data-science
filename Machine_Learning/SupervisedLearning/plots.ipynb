<h1 style="color:green">PieChart</h1>

```python
y=np.array([2,5,3,6,1,8])
mylabels=["apple","samsung","vivo","oppo","lava","google"]
mycolors=np.array(["red","orange","yellow","green","blue","purple"])
myexplode=[0.01,0.02,.029,0.30,0,0]
plt.pie(y,labels=mylabels,explode=myexplode,shadow=True,colors=mycolors)
plt.legend(title="color on products")
plt.show()
```

<h1>read from csv file </h1>

```python
d=pd.read_csv("company_sales_data (1).csv")
plt.plot(d['facecream'],marker='o',linewidth=2,label="facecream")
plt.plot(d['toothpaste'],marker='o',linewidth=2,label="toothpaste")
plt.plot(d['shampoo'],marker='o',linewidth=2,label='shampoo')
plt.plot(d['moisturizer'],marker='o',linewidth=2,label='moisturizer')
plt.legend()
plt.savefig("mygraph.png")
plt.show()
```

```python
x=range(len(d))

plt.bar(x,d['facecream'],label="facecream",width=0.3,color='r',align='center')
# plt.bar([p + 0.2 for p in x ],d['facewash'],label="facewash",width=0.1,color='r')
plt.bar([p + 0.4 for p in x ],d['facecream'],label="facecream",width=0.3,color='b',align='center')
plt.bar([p + 0.6 for p in x ],d['shampoo'],label="shampoo",width=0.3,color='m',align='center')
plt.title("Graph ploted on the basis of line height")
plt.legend()
plt.xlabel("months")
plt.ylabel("sales")
plt.xticks([p + 0.3 for p in x],d['month_number'])
plt.show()
```
